```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
vdf = pd.read_csv("beeActivity.csv")
```

```python
vdf['cy'].max()
```

```python
vdf['cx'].max()
```

```python
vdf = vdf[vdf['group_id'] == 11]
vdf
```

```python
x = vdf['cx']/4
y = vdf['cy']/4.5
im = plt.imread("canvas.png")
implot = plt.imshow(im)
plt.plot(x, y, c="red")
plt.annotate(
    "",
    xy=(x.iloc[-1], y.iloc[-1]),        
    xytext=(x.iloc[-2], y.iloc[-2]),   
    arrowprops=dict(
        arrowstyle="->",
        color="red",
        linewidth=2,
    ),
)

plt.savefig(
    "trajectory.png",
    dpi=300,
    bbox_inches="tight",
    pad_inches=0,
)

```
